```python
# build a lattice model
length = 10
database_dir = "%s/database" % os.getcwd()
c = Conformations(length, database_dir)
s1, s2 = search_conformation_space(c, 1, 5, differby=6)
print(s1, s2)
```

```python
s1 = 'KYAWSLWWKR'
s2 = 'KPDWCGLWIR'
```

```python
z = c.FoldSequence(s1, temp=1)
conf = z[1]
```

```python
ss = LatticeConformationSpace(s1, s2, c)
genotypes = ss.sequences
phenotypes = ss.stabilities
local = LocalEpistasisModel(s1, genotypes, phenotypes)
local.estimate_interactions()
fig,ax = epistasis_bar(local, string_labels=True, figsize=[9,3], color='k',title="Estimated epistasis in protein lattice sequence space")
#fig.savefig("figures/lattice_epistasis2.png", dpi=300, bbox_inches='tight')
```

```python
ss = LatticeConformationSpace(s1, s2, c, target_conf=conf)
genotypes = ss.sequences
phenotypes = ss.stabilities
local = LocalEpistasisModel(s1, genotypes, phenotypes)
local.estimate_interactions()
fig,ax = epistasis_bar(local, string_labels=True, figsize=[9,3], color='k',title="Estimated epistasis in protein lattice sequence space")
#fig.savefig("figures/lattice_epistasis_conf2.png", dpi=300, bbox_inches='tight')
```
