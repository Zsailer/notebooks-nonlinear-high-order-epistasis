```python
# build a lattice model
length = 10
database_dir = "%s/database" % os.getcwd()
c = Conformations(length, database_dir)
s1, s2 = search_conformation_space(c, 1, 5, differby=6)
print(s1, s2)
```

```python
s1 = 'AYCTYDMKNV'
s2 = 'AGIWYDHKNV'
```

```python

ss = LatticeConformationSpace(s1, s2, c)
genotypes = ss.sequences
phenotypes = abs(ss.stabilities)
local = LocalEpistasisModel(s1, genotypes, phenotypes, log_phenotypes=True)
local.estimate_interactions()
fig, ax = epistasis_bar(local, string_labels=True, figsize=figsize2)
```
