# High order epistasis in yeast haploid growth rates. 

Initial imports. Download the `epistasis` package [here](https://www.github.com/harmslab/epistasis)

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import json 

# Epistasis imports
from epistasis.models import LocalEpistasisModel
from epistasis.models.pca import EpistasisPCA
from epistasis.plotting import epistasis_bar, bar_with_xbox
```

```python
in_path = "../../Dropbox/epistasis-paper/datasets/hall_haploid_growth.json"
out_path = "../../Dropbox/epistasis-paper/figures/"
```

Initialize and fit epistasis model to log transformed data, following e

```python
title = "High order epistasis interactions in yeast diploid growth"
model = LocalEpistasisModel.from_json(in_path)
model.fit()
model.fit_error()
```

```python
fig, ax = bar_with_xbox(model,sigmas=2, y_axis_name="change in haploid growth rate",height_ratio=9,figsize=(7,8),star_spacer=0.01)
fig.savefig(out_path + "epistasis-hall-haploid-growth.svg", format="svg", dpi=300, bbox_inches='tight')
```
