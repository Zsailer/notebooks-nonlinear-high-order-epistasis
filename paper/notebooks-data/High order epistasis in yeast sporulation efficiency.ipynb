# High order epistasis in yeast mating efficiency

Initial imports. Download the `epistasis` package [here](https://www.github.com/harmslab/epistasis)

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

# Epistasis imports
from epistasis.models import LocalEpistasisModel
from epistasis.plotting import epistasis_bar, bar_with_xbox
```

Load data from `.json` and `.txt` file.

```python
in_path = "../../Dropbox/epistasis-paper/datasets/hall_sporulation_efficiency.json"
out_path = "../../Dropbox/epistasis-paper/figures/"
```

```python
title = "High order epistasis interactions in yeast sporulation efficiency"
model = LocalEpistasisModel.from_json(in_path)
model.fit()
model.fit_error()
```

```python
fig, ax = bar_with_xbox(model,y_axis_name="change in sporulation efficiency",height_ratio=9,figsize=(7,8),star_spacer=0.01)
fig.savefig(out_path + "epistasis-hall-mating-efficiency.svg", format="svg", dpi=300, bbox_inches='tight')
```
